### مراحل الگوریتم PSO


1. جمعیت پرندگان را با موقعیت ها و سرعت های تصادفی اولیه کنید.
2. تناسب هر پرنده را با محاسبه مقدار تابع در موقعیت فعلی آن ارزیابی کنید.
3. در صورت بهبود تناسب اندام، بهترین راه حل شخصی را برای هر پرنده به روز کنید.
4. به روز رسانی بهترین راه حل جهانی برای کل جمعیت بر اساس تناسب اندام همه پرندگان.
5. سرعت و موقعیت هر پرنده را با استفاده از معادلات PSO به روز کنید.
6. مراحل 2-5 را تکرار کنید تا زمانی که یک معیار توقف، مانند حداکثر تعداد تکرارها یا سطح مطلوب تناسب اندام برآورده شود.
7. در هر تکرار، در صورتی که بهترین راه حل جهانی بهبود یابد، بهترین راه حلی که تاکنون پیدا شده است را به روز کنید.
8. چاپ جهانی و بهترین راه حل های یافت شده توسط الگوریتم PSO.





## introduction !:

مسئله فروشنده مسافر (TSP) یک مسئله بهینه سازی کلاسیک در علوم کامپیوتر است، که در آن هدف یافتن کوتاه ترین مسیر ممکن است که از همه شهرها در یک مجموعه معین بازدید می کند و به شهر شروع باز می گردد. این یک مسئله NP-hard است، به این معنی که هیچ الگوریتم شناخته شده ای وجود ندارد که بتواند آن را در زمان چند جمله ای برای همه نمونه های مسئله حل کند. بهینه‌سازی ازدحام ذرات (PSO) یک الگوریتم بهینه‌سازی فراابتکاری است که از رفتار گله‌های پرندگان یا ماهی‌ها الهام گرفته شده است. در PSO، جمعیتی از ذرات در یک فضای جستجو حرکت می کنند و هر ذره یک راه حل ممکن برای مشکل را نشان می دهد. الگوریتم تناسب هر ذره را ارزیابی می‌کند (یعنی چقدر راه‌حل خوبی را نشان می‌دهد)، و ذرات به سمت بهترین راه‌حلی که تاکنون پیدا شده، به نام بهترین راه‌حل حرکت می‌کنند.
حال بیایید نگاهی دقیق تر به پیاده سازی الگوریتم PSO برای مشکل TSP در کد ارائه شده بیندازیم:

کلا دوتا ایشالایرد دارم که مقدار دهی های اولیه که استاتیک نیستن رو انجام میدن و یه تفاوت خیلی بنیادی با هم دارن اونم اینه که یکیشون خیلی گریدی میاد پرنده های اولیمونو میسازه اون یکی کاملا رندوم و به طور خیلی خیلی عجیبی این گیریدی که میسازم خوب عمل میکنه ینی حتی از خود تابع بهتر عمل میکنه

intializeDataGreedy و intializeDataRandom به ترتیب با یک الگوریتم حریص یا مقداردهی اولیه تصادفی جمعیت رو مقداردهی اولیه می کنن.

### computeDistansMatrix :
ماتریس فاصله بین تمام جفت شهرها را محاسبه می کند.

### fitnessFunction: :
 فیتنس یک مسیر (راه حل) معین رو 
بر اساس طول آن محاسبه می کند.
  ### pathFitness:
   فتنس هر مسیر رو در جمعیت محاسبه می کنه.
  ### evalBirds : 
  بعد اینکه فیتنس هر مسیر رو محاسبه میکنه میاد چیزم میکه اممم متغییر های گلوبالی که داریم تو الگوریتم رو اپدیت میکنه
  ### mutate :
  جهش را روی یک ذره انجام می دهد تا یک ذره جدید تولید کند. بیایید نگاهی دقیق‌تر به عملکرد هر یک از این روش‌ها بیندازیم:

  ### computeDistansMatrix :
  اسمش روشه دیگه میاد ماتریس فاصله هارو درست میکنهبه طور مکرر نزدیک ترین شهر به شهر فعلی را انتخاب می کند تا زمانی که همه شهرها بازدید شوند. مسیر حاصل برای مقداردهی اولیه جمعیت استفاده می شود.


  توابعی هم که دوتا نوشتم یکیشون فقط همون کاری که اسمشرو انحام میده اون یکی مقدار دهی هم میکنه بقیه متغییر هارو

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt


class PSO(object):
    def __init__(self, populationSize=100 , wInersi=0 , c1 =0.1, c2 =0.1 ,itrantions = 20 , doNotAttentionToItrationNumber = False,expectedAns = 2800 , path="" ):
        self.itrations = itrantions 
        self.num = populationSize   
        self.doNotAttentionToItrationNumber= doNotAttentionToItrationNumber
        self.expectedAns=expectedAns
        self.c1=c1
        self.c2=c2
        self.wInersi=wInersi
       
    
    def inputFile(self , path , greedyPath=True):
        lines = open(path, 'r').readlines()
        assert 'NODE_COORD_SECTION\n' in lines
        index = lines.index('NODE_COORD_SECTION\n')
        data = lines[index + 1:-1]
        tmp = []
        for line in data:
            line = line.strip().split(' ')
            if line[0] == 'EOF':
                continue
            tmpline = []
            for x in line:
                if x == '':
                    continue
                else:
                    tmpline.append(float(x))
            if tmpline == []:
                continue
            tmp.append(tmpline)
        data = tmp
        data = np.array(data)
        data = data[:, 1:]
        # self.intializeDataGreedy(data.copy())
        if greedyPath:self.intializeDataGreedy(data.copy())
        else : self.intializeDataRandom(data.copy())




    def intializeDataGreedy(self , data):
        self.numberOfCities = len(data) 
        self.location = data 
       
        self.distansMatrix = self.computeDistansMatrix(
            self.numberOfCities, self.location) 
       
        self.birds = self.greedyIntialization(self.distansMatrix, num_total=self.num, numberOfCities=self.numberOfCities)  
        self.lenths = self.pathFitness(self.birds)  
       
        initL = min(self.lenths)
        initIndex = self.lenths.index(initL)
        initPath = self.birds[initIndex]
       
        init_show = self.location[initPath]
       
        self.localBest = self.birds   
        self.localBest_len = self.lenths   
       
        self.global_best = initPath        
        self.global_best_len = initL       
       
        self.best_l = self.global_best_len      
        self.best_path = self.global_best      

       
        self.iter_x = [0]
        self.iter_y = [initL] 


    def intializeDataRandom(self , data):
        self.numberOfCities = len(data) 
        self.location = data 
       
        self.distansMatrix = self.computeDistansMatrix(
            self.numberOfCities, self.location) 
       
        self.birds = [list(random.sample(range(self.numberOfCities), self.numberOfCities)) for i in range(self.num)]
        self.lenths = self.pathFitness(self.birds)  
       
        initL = min(self.lenths)
        initIndex = self.lenths.index(initL)
        initPath = self.birds[initIndex]
       
        init_show = self.location[initPath]
       
        self.localBest = self.birds   
        self.localBest_len = self.lenths   
       
        self.global_best = initPath        
        self.global_best_len = initL       
       
        self.best_l = self.global_best_len      
        self.best_path = self.global_best      

       
        self.iter_x = [0]
        self.iter_y = [initL] 



   
    def randomInitialization(self, num_total, numberOfCities):
        tmp = [x for x in range(self,numberOfCities)]
        result = []
        for i in range(num_total):
            random.shuffle(tmp)
            result.append(tmp.copy())
        return result

   
    def computeDistansMatrix(self, numberOfCities, location):
        distansMatrix = np.zeros((numberOfCities, numberOfCities))
        for i in range(numberOfCities):
            for j in range(numberOfCities):
                if i == j:
                    distansMatrix[i][j] = np.inf
                    continue
                a = location[i]
                b = location[j]
                tmp = np.sqrt(sum([(x[0] - x[1]) ** 2 for x in zip(a, b)]))
                distansMatrix[i][j] = tmp
        return distansMatrix

   
    def fitnessFunction(self, path, distansMatrix):
        a = path[0]
        b = path[-1]
        result = distansMatrix[a][b]
        for i in range(len(path) - 1):
            a = path[i]
            b = path[i + 1]
            result += distansMatrix[a][b]
        return result

   
    def pathFitness(self, paths):
        result = []
        for one in paths:
            length = self.fitnessFunction(one, self.distansMatrix)
            result.append(length)
        return result

   
    def greedyIntialization(self, distansMatrix, num_total, numberOfCities):
        start_index = 0
        result = []
        for i in range(num_total):
            rest = [x for x in range(0, numberOfCities)]
           
            if start_index >= numberOfCities:
                start_index = np.random.randint(0, numberOfCities)
                result.append(result[start_index].copy())
                continue
            current = start_index
            rest.remove(current)
           
            result_one = [current]
            while len(rest) != 0:
                tmp_min = math.inf
                tmp_choose = -1
                for x in rest:
                    if distansMatrix[current][x] < tmp_min:
                        tmp_min = distansMatrix[current][x]
                        tmp_choose = x

                current = tmp_choose
                result_one.append(tmp_choose)
                rest.remove(tmp_choose)
            result.append(result_one)
            start_index += 1
        return result

    def evalBirds(self):
        min_lenth = min(self.lenths)
        min_index = self.lenths.index(min_lenth)
        cur_path = self.birds[min_index]
       
        if min_lenth < self.global_best_len:
            self.global_best_len = min_lenth
            self.global_best = cur_path
       
        for i, l in enumerate(self.lenths):
            if l < self.localBest_len[i]:
                self.localBest_len[i] = l
                self.localBest[i] = self.birds[i]

   
    def cross(self, cur, best):
        one = cur.copy()
        l = [t for t in range(self.numberOfCities)]
        t = np.random.choice(l, 2)
        x = min(t)
        y = max(t)
        cross_part = best[x:y]
        tmp = []
        for t in one:
            if t in cross_part:
                continue
            tmp.append(t)
       
        one = tmp + cross_part
        l1 = self.fitnessFunction(one, self.distansMatrix)
        one2 = cross_part + tmp
        l2 = self.fitnessFunction(one2, self.distansMatrix)
        if l1 < l2:
            return one, l1
        else:
            return one, l2

   

    def mutate(self, one):
        one = one.copy()
        l = [t for t in range(self.numberOfCities)]
        t = np.random.choice(l, 2)
        x, y = min(t), max(t)
        one[x], one[y] = one[y], one[x]
        l2 = self.fitnessFunction(one, self.distansMatrix)
        return one, l2

   
    def pso(self):
        iteration = 0
        while( self.best_l> self.expectedAns): 
           iteration +=1
           if (not self.doNotAttentionToItrationNumber ):
            if (iteration > self.itrations) : break
            for i, one in enumerate(self.birds):
                tmp_l = self.lenths[i]
               
                new_one, new_l = self.cross(one, self.localBest[i])
                if new_l < self.best_l:
                    self.best_l = tmp_l
                    self.best_path = one

                if new_l < tmp_l or np.random.rand() < self.c1:
                    one = new_one
                    tmp_l = new_l

               
                new_one, new_l = self.cross(one, self.global_best)

                if new_l < self.best_l:
                    self.best_l = tmp_l
                    self.best_path = one

                if new_l < tmp_l or np.random.rand() < self.c2:
                    one = new_one
                    tmp_l = new_l
                
                one, tmp_l = self.mutate(one)

                if new_l < self.best_l:
                    self.best_l = tmp_l
                    self.best_path = one

                if new_l < tmp_l or np.random.rand() < self.c1:
                    one = new_one
                    tmp_l = new_l

               
                self.birds[i] = one
                self.lenths[i] = tmp_l
           
            self.evalBirds()
           
            if self.global_best_len < self.best_l:
                self.best_l = self.global_best_len
                self.best_path = self.global_best
            print(iteration, self.best_l)
            self.iter_x.append(iteration)
            self.iter_y.append(self.best_l)
        return self.best_l, self.best_path

    def run(self):
        best_length, best_path = self.pso()
       
        return self.location[best_path], best_length



```
